## Q1.

Among Decision Tree, KNN, and Linear Regression, the **Decision Tree** algorithm is the most suitable because it handles classification tasks effectively and can make decisions based on multiple conditions such as income, credit score, and employment status.

## Q2.

Explain the difference between a feature and a label using a real-world example.

**Answer:**
A **feature** is an input variable used to make a prediction, while a **label** is the output that the model tries to predict.

For example, in a house price prediction system:

* Features: Area of the house, number of bedrooms, location, and age of the house.
* Label: House price.

The model uses the features to predict the label.

## Q3.

A Decision Tree predicts that a customer will purchase a product. Explain how the model reaches this prediction.

**Answer:**
A Decision Tree makes predictions by following a series of decision rules. Starting from the root node, it checks conditions such as customer age, income, or previous purchases. Based on the answers to these conditions, the model moves through different branches until it reaches a leaf node. The class stored in the leaf node becomes the final prediction. In this case, the leaf node indicates that the customer is likely to purchase the product.

## Q4.

Why is KNN considered a supervised learning algorithm? Can KNN perform classification if labels are not available?

**Answer:**
KNN is a supervised learning algorithm because it learns from training data that contains both features and labels. When making a prediction, it compares a new data point with labeled examples.

No, KNN cannot perform classification if labels are not available because it relies on the labels of neighboring data points to determine the predicted class.

## Q5.

Suppose a KNN model uses K = 5 and the nearest neighbors are:
Yes, Yes, No, Yes, No

**Answer:**
The prediction will be **Yes**.

This is because KNN uses majority voting. Among the five nearest neighbors:

* Yes = 3
* No = 2

Since "Yes" appears more frequently, the model predicts "Yes".

## Q6.

Explain the equation of Linear Regression:

y = mx + b

What do x, y, m, and b represent?

**Answer:**
In the equation y = mx + b:

* x = Independent variable (input feature)
* y = Dependent variable (predicted output)
* m = Slope of the line, showing how much y changes when x changes
* b = Intercept, representing the value of y when x = 0

This equation describes the relationship between the input and output variables.

## Q7.

What is correlation? What does each of the following values indicate?

+1, 0, -1

**Answer:**
Correlation measures the strength and direction of the relationship between two variables.

* +1: Perfect positive correlation. As one variable increases, the other also increases.
* 0: No correlation. There is no linear relationship between the variables.
* -1: Perfect negative correlation. As one variable increases, the other decreases.

## Q8.

A disease detection model produced the following result:

Actual Condition: Disease Present

Prediction: Healthy

What type of prediction is this? Why can this be dangerous?

**Answer:**
This is a **False Negative** prediction.

It is dangerous because the model incorrectly predicts that a sick patient is healthy. As a result, the patient may not receive proper treatment, which could worsen the disease and lead to serious health complications.

## Q9.

What is the purpose of a Confusion Matrix? How is it different from a Correlation Matrix?

**Answer:**
A **Confusion Matrix** is used to evaluate the performance of a classification model. It shows the number of True Positives, True Negatives, False Positives, and False Negatives.

A **Correlation Matrix** is used to measure relationships between numerical variables. It displays correlation values ranging from -1 to +1.

Therefore, a Confusion Matrix evaluates classification results, while a Correlation Matrix analyzes relationships between variables.
